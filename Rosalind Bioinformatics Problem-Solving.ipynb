# Counting DNA Nucleotides

```python
from collections import Counter

def NucCount(string):
    Ac = 0
    Cc = 0
    Gc = 0
    Tc = 0
    for letter in string:
        if letter == 'A':
            Ac += 1
        if letter == 'C':
            Cc += 1
        if letter == 'G':
            Gc += 1
        if letter == 'T':
            Tc += 1
    
    return (str(Ac) + " " + str(Cc) + ' ' + str(Gc) + ' ' + str(Tc))

f=open("Downloads/rosalind_dna.txt", "r")
if f.mode == 'r':
    contents =f.read()
    
print NucCount(contents)

print Counter(contents)
```

# Complementing a Strand of DNA

```python
def Complement(string):
    reverse = string[::-1]
    reverse = reverse.replace("A","Ax")
    reverse = reverse.replace("C","Cx")
    reverse = reverse.replace("G","C")
    reverse = reverse.replace("T","A")
    reverse = reverse.replace("Ax","T")
    reverse = reverse.replace("Cx","G")
    return reverse

f=open("Downloads/rosalind_revc.txt", "r")
if f.mode == 'r':
    contents =f.read()
    
    
print Complement(contents)
```

# Mendel's First Law

```python
def MendProbability(k,m,n):

    k = float(k)
    m = float(m)
    n = float(n)
    total = k+m+n
    P1 = (n/total) * ((n-1)/(total-1))
    P2 = (m/total) * ((m-1)/(total-1)) * .25
    P3 = (m/total) * ((n)/(total-1)) * .5
    P4 = (n/total) * ((m)/(total-1)) * .5

    P = 1 - (P1+P2+P3+P4)

    return P

int_input=open("Downloads/rosalind_iprb.txt", "r").read().split()
print int_input
k,m,n = int_input[0], int_input[1], int_input[2]
f.close()


print MendProbability(k,m,n)
```

# Consensus and Profile

When you have multiple homologous strands, you can analyze by finding an average-case strand to represent the most likely common ancestor of the given strands

```python
from Bio import SeqIO
from __future__ import print_function
fas_dict={}
file_path = 'Downloads/rosalind_cons.txt'
with open(file_path, mode='r') as handle:
    for record in SeqIO.parse(handle, 'fasta'):
        fas_dict.update({record.id:record.seq})
        

def ConsensusProfile(fas_dict):
    
    n = len(fas_dict)
    m = len(fas_dict.values()[0])
    profile = {'A':[0]*m,
               'C':[0]*m,
               'G':[0]*m,
               'T':[0]*m}
    consensus = ''

    for i in range(m):
        for j in range(n):
            profile[fas_dict.items()[j][1][i]][i] += 1
   
    for i in range(m):
        best_letter = ''
        best_score = 0
        for j in "ACGT":
            if profile[j][i] > best_score:
                best_score = profile[j][i]
                best_letter = j
        consensus += best_letter
    
    #Output changed to match rosalind required formating
    print (*profile.values()[0], sep=' ')
    print (*profile.values()[1], sep=' ')
    print (*profile.values()[3], sep=' ')
    print (*profile.values()[2], sep=' ')
    return consensus, profile


print (ConsensusProfile(fas_dict))
```

# Mortal Fibonacci Rabbits

```python
def TotalMortalRabbitPairs(n,m):
    n = int(n)
    m = int(m)
    F = []
    F.append(1)
    F.append(1)
    
    for i in range(2, n):
        if i < m:
            F.append((F[i-1] + F[i-2]))
        elif i == m:
            F.append((F[i-1] + F[i-2]) - (F[i-m]))
        else:
            F.append((F[i-1] + F[i-2]) - (F[i-m-1]))
    
    return F[n-1]

file_path = 'Downloads/rosalind_fibd.txt'
data = [x for x in open(file_path).read().split(" ")]
n, m = data[0], data[1]
#n, m = 6,3

print (n)
print (m)
print (TotalMortalRabbitPairs(n,m))
```

# Calculating Expected Offspring

```python
#p1 = AA-AA
#p2 = AA-Aa
#p3 = AA-aa
#p4 = Aa-Aa
#p5 = Aa-aa
#p6 = aa-aa

def ExpectedOffspring(p1,p2,p3,p4,p5,p6):
    total = (2*(p1+p2+p3)) + 1.5*(p4) + (p5)
    return total

file_path = 'Downloads/rosalind_iev.txt'
value_list = [x for x in open(file_path).read().split(" ")]
p1 = float(value_list[0])
p2 = float(value_list[1])
p3 = float(value_list[2])
p4 = float(value_list[3])
p5 = float(value_list[4])
p6 = float(value_list[5])

print(ExpectedOffspring(p1,p2,p3,p4,p5,p6))
    
    
```

# Finding a Shared Motif

```python
from Bio import SeqIO
import re

def SharedMotifs(name, sequence):
    
    longest_seq = max(sequence, key = len)
    n= len(longest_seq)
    shared_motif = ''
    sequence.remove(longest_seq)
    bool1 = False
    
    for i in range(2,n+1):
        for j in range (0,n-i+1):
            for seq in sequence:
                if longest_seq[j:j+i] in seq:
                    bool1 = True
                else:
                    bool1 = False
                    break
            if bool1 == True:
                shared_motif = longest_seq[j:j+i]

    
    return shared_motif

file_path = 'Downloads/rosalind_lcsm.txt'
name = []
sequence = []
fasta_sequences = SeqIO.parse(open(file_path),'fasta')
for fasta in fasta_sequences:
        name.append(fasta.id)
        sequence.append(str(fasta.seq))
        
print(SharedMotifs(name, sequence))
```

# Independent Alleles

```python
#independent event probabilities can be calculated separately and then multiplied to find the prob of both occuring
#
#Start with Tom -> has Aa Bb in the 0th generation
#has 2 children in 1st gen and each child has 2 children and so on..
#Each organism mates with Aa Bb..
#
#Return the probability that at least N(given) Aa Bb organisms will belong to the k-th(given) generation
import math


#math.factorial(n)/(math.factorial(n-i)*math.factorial(n-(n-i)))
def IndProbability(k,N):
    n = 2**k
    NotP = 0
    for i in range(int(n)):
        if N == i:
            break
        else:
            NotP += (0.75**(n-i))*((0.25**i)*(math.factorial(n)/(math.factorial(n-i)*math.factorial(n-(n-i)))))
    
    P = 1 - NotP
    return P

file_path = 'Downloads/rosalind_lia.txt'
data = open(file_path).read().split(' ')
k, N = data[0], data[1]



P = 1-NotP
print(IndProbability(float(k),float(N)))
```
